```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
df = pd.read_csv("data/Churn_Modelling.csv")
```

```python
df.head(5)
```

```python
df.info()
```

```python
df.drop(columns = ["CustomerId", "RowNumber", "Surname"], axis = 1, inplace = True)
```

```python
df.head(5)
```

```python
df.Age.min()
```

```python
df.Age.max()
```

```python
labels = ["0-20", "21-40", "41-60", "Above 61"]
bins = [0, 20, 40, 60, 100]
#The pd.cut() function in Pandas is used to bin (or categorize) continuous data into discrete intervals or categories.
df["Age_Bins"] = pd.cut(df.Age, bins, labels = labels, include_lowest = True)
```

```python
df.head(5)
```

```python
df.Age_Bins.value_counts()
```

```python
# making bar chart on data
plt.bar(labels, df.Age_Bins.value_counts())

# giving the title
plt.title("Age Count")

# giving X and Y labels
plt.xlabel("Age Bins")
plt.ylabel("Age Count")

# visualizing the plot
plt.show()
```
