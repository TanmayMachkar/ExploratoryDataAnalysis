```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
df = pd.read_csv("data/Churn_Modelling.csv")
```

```python
df.head(5)
```

```python
df.info()
```

```python
#When you use df.drop() with axis=1, you are removing columns based on their labels.
df.drop(columns = ["CustomerId", "RowNumber", "Surname"], axis = 1, inplace = True)
```

```python
df.head(5)
```

```python
### Handling missing values in Gender

df.Gender.mode()
```

```python
df["Gender"] = df["Gender"].fillna("Male")
```

```python
### Label Encoding

le = preprocessing.LabelEncoder()
df["Gender_label"] = le.fit_transform(df.Gender.values)
```

```python
df.head(5)
```

```python
df.Gender_label.value_counts()
```

```python
### One Hot Encoding

one_hot = pd.get_dummies(df["Geography"])
```

```python
one_hot
```

```python
df_dummies = pd.get_dummies(df)

df_dummies.head(5)
```

```python
### Dummy Encoding

df_dummies_de = pd.get_dummies(df, drop_first = True)
df_dummies_de.head(5)
```
